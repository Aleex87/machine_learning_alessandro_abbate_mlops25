## Short EDA on ads Data

```python
import pandas as pd 
df = pd.read_csv("Advertising.csv", index_col=0)
```

```python
type(df)
type(df["TV"])
df.info()
```

- features: TV, radio, newspaper
- label/target/dependent variable: sales

```python
import seaborn as sns

sns.scatterplot(data = df, x="TV", y="sales")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1,3, figsize=(14,5))
sns.scatterplot(data= df, x="TV", y="sales", ax=axes[1])
sns.scatterplot(data= df, x="radio", y="sales", ax=axes[2])
sns.scatterplot(data= df, x="newspaper", y="sales", ax=axes[0])

axes[0].set(title= "Newspapaer spending Vs sales")
axes[1].set(title= "TV spending Vs sales")
axes[2].set(title= "Radio spending Vs sales")
```

- did the same thing but another way to do it with a foor loop

```python
fig,axes = plt.subplots(1,3, figsize=(14,5))

features = ("TV", "radio", "newspaper")
for feature, ax in zip(features, axes.flatten()):
 sns.scatterplot(data= df, x=feature, y="sales", ax=ax)
 ax.set(title=f"{feature} Vs sales")
```

- We have rapresented all the data compared with sales (label colum) using the matplotlib.pyplot library
